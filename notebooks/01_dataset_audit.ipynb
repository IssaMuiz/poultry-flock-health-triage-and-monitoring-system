```python
from pathlib import Path
import random
from PIL import Image
import matplotlib.pyplot as plt
```

```python
DATASET_PATH = Path("../data/raw/Poultry Birds Poo Imagery Dataset for Health Status Prediction A Case of South-West Nigeria/Dataset")
```

```python
class_folders = sorted([
    folder for folder in DATASET_PATH.iterdir() if folder.is_dir() and not folder.name.startswith(".")
    ])

[class_folder.name for class_folder in class_folders]
```

```python
fig, axes = plt.subplots(
    nrows=2,
    ncols=5,
    figsize=(15, 6)
)

for row, class_folder in enumerate(class_folders):

    image_paths = list(class_folder.glob("*"))

    random_images = random.sample(image_paths, 5)

    for col, image_path in enumerate(random_images):

        image = Image.open(image_path)

        axes[row, col].imshow(image)

        axes[row, col].set_title(class_folder.name)

        axes[row, col].axis("off")

plt.tight_layout()

plt.show()
```
